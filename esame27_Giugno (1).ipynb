## Esercizio 3

Crea una funzione di nome **mat_normalizzata** che prende come input una matrice e ritorna la matrice normalizzata.

La funzione deve quindi:
1. Calcolare la media e la deviazione standard della matrice.
2. Normalizzare la matrice sottraendo la media e dividendo per la deviazione standard.

Testa La funzione con un array NumPy 6x10 con numeri campionati da una distribuzione normale (media=5, std=2).

```python
#1- 
#2-
def mat_normalizzata(matrix):
    med = matrix.mean(axis=0)
    std = matrix.std(axis=0)
    matrix = (matrix - med) / std
    return matrix

vet = np.random.normal(5,2,size=(6,10))
mat_normalizzata(vet)
```

## Esercizio 6

Genera due array:

`x = np.linspace(0, 5, 80)`  
`y = 2.5 * x + 1.2 + np.random.normal(0, 0.5, 80)`

1. Usa `scipy.optimize.curve_fit` per stimare i parametri della retta.
2. Plotta i punti originali e la retta ottenuta.
3. Calcola MAE e RMSE tra i valori reali e quelli stimati.

```python
from scipy.optimize import curve_fit

x = np.linspace(0, 5, 80)
y = 2.5 * x + 1.2 + np.random.normal(0, 0.5, 80)

#1-
def model(x,a,b):
    return a*x + b

params, variance = curve_fit(model, x, y)
a,b=params
a = round(a,3)
b = round(b,3)
print("Abbiamo quindi i parametri a = {} e b = {}".format(a,b))
```

```python

#2-
y_predite = model(x, a, b)
sns.scatterplot(x=x, y=y, label="Dati originali")
sns.lineplot(x=x, y=y_predite, color="red", label="punti prediti")
plt.xlabel("valore x")
plt.ylabel("valore y")
plt
plt.show()
```

```python

#3-
MAE = np.sum(np.abs(y-y_predite)) / len(y)
RMSE = np.sqrt(np.sum((y-y_predite)**2))
MAE = round(MAE,3)
RMSE= round(RMSE,3)
print("il MAE vale {} e il RMSE vale {}".format(MAE, RMSE))
```
